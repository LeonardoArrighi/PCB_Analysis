## Plot

In this notebook we plot the results of the training.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
name = 'C:\\Users\\leonardo.arrighi\\Documents\\Asac\\TEST_14_low_viz\\slurm-9312759.out'
```

```python
f = open(name, 'r')
b = [i for i in f]
print(b[0])
print(b[0][-85:-79])
f.close()
```

```python
f = open(name, 'r')
tl = [float(i[-85:-79]) for i in f]
f.close()

plt.figure(figsize=(16, 10))
plt.ylabel('Train loss')
plt.xlabel('Number of iterations')
plt.title('Train loss vs number of iterations')
plt.plot([i for i in range(len(tl))], tl, 'b')
```

```python
f = open(name, 'r')
vl = [float(i[-36:-30]) for i in f]
f.close()

plt.figure(figsize=(16, 10))
plt.ylabel('Validation loss')
plt.xlabel('Number of iterations')
plt.title('Validation loss vs number of iterations')
plt.plot([i for i in range(len(vl))], vl, 'r')
```

```python
plt.figure(figsize=(20, 12))
plt.ylabel('Loss', size = 30)
plt.xlabel('Number of iterations', size = 30)
# plt.title('Train / Validation loss vs number of iterations', size = 25)
plt.plot([i for i in range(len(vl[0:300]))], vl[0:300], 'r-', label = 'Validation loss')
plt.plot([i for i in range(len(tl[0:300]))], tl[0:300], 'b-', label = 'Train loss')
plt.legend(prop={'size': 30})
```

```python
f = open(name, 'r')
ta = [float(i[-61:-55]) for i in f]
f.close()

plt.figure(figsize=(16, 10))
plt.ylabel('Train accuracy')
plt.xlabel('Number of iterations')
plt.title('Train accuracy vs number of iterations')
plt.plot([i for i in range(len(ta))], ta, 'b')
```
